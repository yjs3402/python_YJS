```python
import pandas as pd
import folium

#    
df = pd.read_excel('./data/__.xlsx')

#   
seoul_map = folium.Map(location=[37.55, 126.98], tiles='Cartodb Positron',
                       zoom_start=12)

#   Marker 
for name, lat, lng in zip(df.index, df[''], df['']):
    folium.Marker([lat, lng], popup=name).add_to(seoul_map)

#   
seoul_map
```
