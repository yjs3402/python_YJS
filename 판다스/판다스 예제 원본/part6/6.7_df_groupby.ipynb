```python
# 라이브러리 불러오기
import pandas as pd
import seaborn as sns

# titanic 데이터셋에서 age, sex 등 5개 열을 선택하여 데이터프레임 만들기
titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age','sex', 'class', 'fare', 'survived']]

df.head()
```

```python
# class 열을 기준으로 분할
grouped = df.groupby(['class'], observed=True) 
print(grouped)
```

```python
# 그룹 객체를 iteration으로 출력: head() 메소드로 첫 5행만을 출력
for key, group in grouped:
    print('* key :', key)
    print('* number :', len(group))    
    print(group.head())
    print('\n')
```

```python
# 연산 메소드 적용
average = grouped.mean(numeric_only=True)
average
```

```python
# 개별 그룹 선택하기
group3 = grouped.get_group(('Third',))
group3.head()
```

```python
# class 열, sex 열을 기준으로 분할
grouped_two = df.groupby(['class', 'sex'], observed=True) 

# grouped_two 그룹 객체를 iteration으로 출력
for key, group in grouped_two:
    print('* key :', key)
    print('* number :', len(group))    
    print(group.head())
    print('\n')
```

```python
# grouped_two 그룹 객체에 연산 메소드 적용
average_two = grouped_two.mean(numeric_only=True)
average_two
```

```python
# grouped_two 그룹 객체에서 개별 그룹 선택하기
group3f = grouped_two.get_group(('Third','female'))
group3f.head()
```
