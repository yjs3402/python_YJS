```python
# 라이브러리 불러오기
import seaborn as sns

# titanic 데이터셋 가져오기
df = sns.load_dataset('titanic')

# 첫 5행 출력
df.head()
```

```python
# 데이터프레임 개요
df.info()
```

```python
# deck 열의 NaN 개수 계산하기
df['deck'].value_counts(dropna=False) 
```

```python
# isnull() 메서드로 누락 데이터 찾기
df.head().isnull()
```

```python
# notnull() 메서드로 누락 데이터 찾기
df.head().notnull()
```

```python
# isnull() 메서드로 누락 데이터 개수 구하기
df.head().isnull().sum(axis=0)
```

```python
# missingno 라이브러리 활용
import missingno as msno
import matplotlib.pyplot as plt

# 매트릭스 그래프
msno.matrix(df)
plt.show()
```

```python
# 막대 그래프
msno.bar(df)
plt.show()
```

```python
# 히트맵 
msno.heatmap(df)
plt.show()
```

```python
# 덴드로그램 
msno.dendrogram(df)
plt.show()
```
