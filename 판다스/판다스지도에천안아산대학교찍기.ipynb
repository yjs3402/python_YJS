```python
import pandas as pd
import folium

loca = {'위도': ['36.8512476', '36.7989764', '36.7907078', '36.7962041', '36.7708159', '36.8374266', '36.8394518', '36.8336012', '36.909155', '36.9501075'],
        '경도': ['127.0615552', '127.0750025', '127.1075771', '127.1185261', '127.1643821', '127.1687229', '127.1839014', '127.1791657', '127.144395', '127.1557029']}
index  = ['유원대학교', '선문대학교 아산', '호서대학교', '나사렛대학교', '선문대학교 천안', '단국대학교', '백석대학교', '상명대학교', '남서울대학교', '연암대학교']
df1 = pd.DataFrame(loca, index=index)
df1
```

```python
# CSV 파일로 내보내기
df1.to_csv('./data/df_sample.csv')
```

```python
df2 = pd.read_csv('./data/df_sample.csv')
df2 = df2.set_index('Unnamed: 0')
# 천안 지도
cheonan_map = folium.Map(location=[36.8135845, 127.1416895], tiles="Cartodb Positron",
                         zoom_start=12)

# 대학교 위치정보를 Marker로 표시
for name, lat, lng in zip(df2.index, df2['위도'], df2['경도']):
    folium.Marker([lat, lng], popup=name).add_to(cheonan_map)

# 지도 객체 확인
cheonan_map
```
