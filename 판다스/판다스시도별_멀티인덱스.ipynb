```python
# 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt
from math import ceil

# matplotlib 한글 폰트 오류 문제 해결
from matplotlib import font_manager, rc
font_path = "../part4/data/malgun.ttf"   #폰트파일의 위치
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# Excel 데이터를 데이터프레임 변환 
df = pd.read_excel('./data/시도별_전출입_인구수.xlsx')

# NaN 값을 바로 위의 값으로 변환
df = df.ffill()
df = df.loc[1:]
df = df.replace('-', 0)

# 데이터프레임 생략 안되게 설정
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

df_grouped = df.groupby(['전출지별', '전입지별'], observed=True)

gdf = df_grouped.mean(numeric_only=True)
gdf
gdf.loc[('강원도', )]


# for key, group in df_grouped:
#     display(group.head())
#     print('\n')
```

```python
col_years = list(map(str, range(1970, 2018)))
colorss = [ 'brown', 'red', 'coral', 'gold', 'yellow', 'greenyellow', 'yellowgreen', 'darkolivegreen', 'darkgreen', 'lightsteelblue', 'royalblue', 'blue', 'mediumpurple', 'blueviolet', 'purple', 'violet', 'white', 'black']
# 스타일 서식 지정
plt.style.use('ggplot')

# 그래프 객체 생성(figure 에 1개의 서브 플롯을 생성)
fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1, 1, 1)

# axe 객체에 plot 함수로 그래프 출력
ldf = gdf.loc[('강원도', )]
for idx in range(len(ldf.index)):
    print(idx)
    loca = ldf.index[idx]
    ax.plot(col_years, ldf.loc[loca], marker='', markerfacecolor=colorss[idx],
            markersize=10, color=colorss[idx], linewidth=2, label=f'강원도 -> {loca}')
```
